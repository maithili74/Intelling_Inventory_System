```python
import pandas as pd
import numpy as np
```

```python
inventory_df = pd.read_csv(
    r"C:\Users\MAITHILI\Personal_project\inventory_optimization_output.csv"
)

print(inventory_df.head())
```

```python
np.random.seed(42)

inventory_df['Current_Inventory'] = (
    inventory_df['Reorder_Point'] * np.random.uniform(0.5, 1.5, size=len(inventory_df))
).round(0)
```

```python
def inventory_decision(row):
    if row['Current_Inventory'] <= row['Reorder_Point']:
        return pd.Series(['REORDER NOW', row['EOQ']])
    else:
        return pd.Series(['OPTIMAL', 0])
```

```python
inventory_df[['Inventory_Status', 'Order_Quantity']] = inventory_df.apply(
    inventory_decision,
    axis=1
)
```

```python
final_decisions = inventory_df[[
    'Store_ID',
    'Product_ID',
    'ABC_Category',
    'Avg_Daily_Demand',
    'Current_Inventory',
    'Reorder_Point',
    'Safety_Stock',
    'EOQ',
    'Max_Stock_Level',
    'Inventory_Status',
    'Order_Quantity'
]]


print(final_decisions.head(10))
```

```python
final_decisions.to_csv(
    r"C:\Users\MAITHILI\Personal_project\inventory_decision_output.csv",
    index=False
)

print("✓ Inventory decision output saved")
```
